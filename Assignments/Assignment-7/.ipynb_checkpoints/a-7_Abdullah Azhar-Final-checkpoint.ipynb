```python
%reset -f
import numpy as np
from skimage import feature
import matplotlib.pyplot as plt
import cv2
import os
from IPython.display import clear_output
from time import sleep
import time
print('libraries imported!')
```

```python
def Images_Discs(path=""):
    """
    Reading the images and storing them in a dictionary
    # returns a dictionary of images
    """
    directory = os.listdir(path)
    images = {}
    for image in directory:
        I = plt.imread(os.path.join(path,image)) #reads the image 
        images[image] = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY) #converts the image to grey scale
    return images
```

```python
def Salients(images_greys={}):
    #takes a dictionary of grey scale images and returns a dictionary of salients
    Salients_dict = {}
    
    for image in images_greys:
        Salients = feature.canny(image=images_greys[image], sigma=2.5).astype(np.int32)
        Salients[1,:] = 0
        Salients[Salients.shape[0]-2,:] = 0
        Salients[:,1] = 0
        Salients[:,Salients.shape[1]-2] = 0
        
        plt.imshow(Salients,cmap="gray")
        plt.title(f"Salients for: {image} computed")
        plt.show()
        
        Salients_indices = np.argwhere(Salients)
        Salients_dict[image] = Salients_indices
    return Salients_dict #returning dictionary of salients
```

```python
def E_M(I_grey,Salients_indices):
    # main crux of this notebook that computes EM Maximization for the image provided with its corresponding salients
    
    # Initial Conditions:
    Cx = I_grey.shape[0]//2
    Cy = I_grey.shape[1]//2
    sigma = 128
    r = 190
    print(f"Initial conditions initialized as follows: Cx:{Cx}, Cy:{Cy},r:{r}")
    num_iter = 0
    
    Cx_original = Cx #for final plot to compare how the circle changed throughout the iteration
    Cy_original = Cy
    r_original = r
    

    X = Salients_indices[:,0]
    Y = Salients_indices[:,1]
    B = np.vstack(((X**2+Y**2),X,Y,np.ones((X.shape[0])))).T #computing the matrix B for the implementation in gander's section 2
    print(f"Matrix B generated with the following dimensions: {B.shape[0]} rows and {B.shape[1]} columns")
    
    t1 = time.time()
    while True:
        print(f"center: {Cx,Cy}")
        print(f"radius: {r}")
        
        # E-STEP
        d = np.abs(np.sqrt((X-Cx)**2+(Y-Cy)**2)-r) #using the model parameters estimated in the previous loop and
        # computing the new residuals
        
        prob_model_1 = np.exp(-d/sigma)**2 #probability is gaussian in nature
        prob_model_2 = d*(1/10) #probability is uniform in nature 
        denominator = prob_model_1+prob_model_2
        
        #Assuming Equal Priors i.e., every data point initially is equally likely to belong to either one of the two models    
        residual_1 = prob_model_1/denominator
        residual_1 = np.identity((residual_1.shape[0]))*np.reshape(residual_1,(residual_1.shape[0],1)) #setting up
        # the weights matrix with the shape nxn using identity matrix

        #Now that we have our residuals, we need to reupdate the model parameters. Since model 1 is a circle, we 
        # need to reestimate the Cx, Cy, r. Model 2 is an outlier model so no parameters to be updated. 

        # M-STEP
        for_svd_model_1 = np.dot(residual_1,B) #should give us nx4 matrix (weighted B)
        
        u1,s1,v1 = np.linalg.svd(for_svd_model_1) #computing SVD of weighted B matrix
        s1_min = np.where(s1 == s1.min())[0] # now we need to pick this column from the v1 matrix having the minimum value 
        m1_params = v1[s1_min[0],:] # a 1x4 column vector should be returned (a,b1,b2,c)

        Cx_new = -m1_params[1]/(2*m1_params[0])
        Cy_new = -m1_params[2]/(2*m1_params[0])

        c_a = m1_params[3]/m1_params[0] #c/a
        b_mag =  (m1_params[1]**2+m1_params[2]**2)#||b||^2
        r_new = np.sqrt((b_mag/(4*m1_params[0]**2))-(c_a))

        #stopping conditions
        dif_r = r_new-r
        dif_cx = Cx_new - Cx
        dif_cy = Cy_new - Cy

        if dif_r < 0.1 and dif_cx < 0.1 and dif_cy < 0.1:
            print(f"EM converged at {num_iter}th iteration!")
            break
            
        t2 = time.time()
        
        if t2-t1 > 30:
            print(f"Convergence not attained after {t2-t1} seconds so breaking the loop")
        
        r = r_new
        Cy = Cy_new
        Cx = Cx_new
        num_iter += 1

        clear_output(wait=True)
        figure, axes = plt.subplots() 
        cc = plt.Circle((Cy,Cx),r,fill=False) 
        axes.set_aspect(1) 
        axes.add_artist( cc ) 
        axes.set_xlim(0,I_grey.shape[1])
        axes.set_ylim(0,I_grey.shape[1])
        plt.title( f'Estimated Circle at {num_iter}th iteration' ) 
        plt.imshow(I_grey,cmap="gray")
        plt.show()
        sleep(0.5)
    
    
    figure, axes = plt.subplots() 
    cc = plt.Circle((Cy_original,Cx_original),r_original,fill=False) 
    axes.set_aspect(1) 
    axes.add_artist( cc ) 
    axes.set_xlim(0,I_grey.shape[1])
    axes.set_ylim(0,I_grey.shape[1])
    plt.title( f'Circle before EM iteration' ) 
    plt.imshow(I_grey,cmap="gray")
    plt.show()
    
    figure, axes = plt.subplots() 
    cc = plt.Circle((Cy,Cx),r,fill=False) 
    axes.set_aspect(1) 
    axes.add_artist( cc ) 
    axes.set_xlim(0,I_grey.shape[1])
    axes.set_ylim(0,I_grey.shape[1])
    plt.title( f'Estimated Circle at {num_iter}th iteration' ) 
    plt.imshow(I_grey,cmap="gray")
    return None
```

#### Comment on the solution achieved for the fourth image

Since the salients image does not contain the entire periphery of the circle, the EM maximization is confused since the distance of each point from the boundary is not calculated precisely and thus the circle continues to shrink till it converges at the wrong solution. 

Potential solution would be to run EM on the inner salients that is the inner circle of the disc(ignore the boundary of the disc) and once the solution converges, estimate the radius (by comparing 1s in the Salients) and extrapolate the estimated circle
