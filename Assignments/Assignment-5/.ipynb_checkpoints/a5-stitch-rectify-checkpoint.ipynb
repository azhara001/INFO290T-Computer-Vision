```python
import cv2
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
print("Libraries Imported")
```

```python
# transform the image using the homography H 
# the size of the final image is decided using the bounds -> (left, top, right, bottom) 
# the left, top of the input image is translated to (0,0)
def warp_image_with_bounds(im, H, bounds):
    t      = [-bounds[0],-bounds[1]] # translation of left, top to (0,0)
    Ht     = np.array([[1,0,t[0]],[0,1,t[1]],[0,0,1]]) # translation transform
    result = cv2.warpPerspective(im, Ht.dot(H), (bounds[2]-bounds[0], bounds[3]-bounds[1]))
    return result
```

```python
# Reversable handler: allows images to be reversed and still work with the code
def reverse_image(xmin, xmax, ymin, ymax, bounds):
    for x in bounds:
        if x[0][0] > xmax:
            xmax = x[0][0].astype(np.int32)
        if x[0][0] < xmin:
            xmin = x[0][0].astype(np.int32)
        if x[0][1] > ymax:
            ymax = x[0][1].astype(np.int32)
        if x[0][1] < ymin:
            ymin = x[0][1].astype(np.int32)
    return xmin, xmax, ymin, ymax
```

```python
# find the homography between the source and destination images using the specified points
def find_homography(im_src, im_dst):
    # four corresponding points on source and destination image
    pts_src = np.array([[1649.15, 298], [2293.79, 236.60], [2243.19, 2168.90], [1770.98, 2133.52]])
    pts_dst = np.array([[495, 100], [1135, 131], [1021, 1986], [555,1995]])
    
    # display images and points
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
    ax[0].imshow(im_src); ax[0].plot(pts_src[:,0], pts_src[:,1], 'ro', markersize=10); ax[0].axis('off')
    ax[1].imshow(im_dst); ax[1].plot(pts_dst[:,0], pts_dst[:,1], 'ro', markersize=10); ax[1].axis('off')
    plt.show()
    
    # compute homography between src and dst
    # YOUR CODE HERE
    H,ones = cv2.findHomography(pts_src,pts_dst)
    print('Homography:\n',H)
    print(f"\nShape of Homography Matrix: {H.shape}")
    return H
```

## ABDULLAH COMMENTS ON stitch_images function (Assignment Part 2: Bounds)

#### Given two images im_src and im_dst, we want to first compute the bounds of the RESULTANT IMAGE (stitched image). Using the four corners specified in find_homography() function above, we were able to provide the pts_src and their planar mapping in pts_dst. These one to one mappings then gave us H (the homography 3x3 matrix). Now that we have this matrix, we want to input the source image corners and get their mapping in the final image (by using the H matrix)

PS. The width coordinates of the image corresspond to columns so I have set them as xmin, xmax and the height coordinates corresspond to rows so I have set them as ymin,ymax because this code: **h1, w1, d1  = im_src.shape** assumes h1 is the no. of rows and w1 is the no. of columns

```python
def stitch_images(im_src, im_dst):
    # estimate homography
    H = find_homography(im_src, im_dst)
    
    # load images
    h1, w1, d1  = im_src.shape
    h2, w2, d2  = im_dst.shape   
    src_corners = np.float32([[0,0],[0,h1],[w1,h1],[w1,0]]).reshape(-1,1,2) 
    dst_corners = np.float32([[0,0],[0,h2],[w2,h2],[w2,0]]).reshape(-1,1,2)
    
    # FIND BOUNDS OF THE STITCHED IMAGE
    # YOUR CODE HERE
    # [xmin, xmax, ymin, ymax] = ...
    
    src_corners_warped = cv2.perspectiveTransform(src_corners,H)
    minimas = np.minimum(src_corners_warped,dst_corners) # to compute minimums from the destination 
    #image and the warped source image
    
    maximas = np.maximum(src_corners_warped,dst_corners) # to compute maximums from the destination 
    #image and the warped source image
        
    [xmin,xmax,ymin,ymax] = [np.min(minimas[:,:,0]), np.max(maximas[:,:,0]),np.min(minimas[:,:,1]),np.max(maximas[:,:,1])]
    
    # warp the image with bounds
    result = warp_image_with_bounds(im_src, H, [xmin, ymin, xmax, ymax])
    
    # place the destination image into the result
    # YOUR CODE HERE
    width_translation,height_translation = 615,result.shape[1]-im_dst.shape[1] #615 value was fixed via trial and error
    
    for channel in range(im_dst.shape[2]):
        for row in range(im_dst.shape[0]):
            for column in range(im_dst.shape[1]):
                result[width_translation+row,height_translation+column,channel] = im_dst[row,column,channel]
    
    
    # visualize the result
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))
    ax.imshow(result)
    ax.axis('off')
    
    return result
```

```python
def rectify(im_src):    
    # four corresponding points on source and destination image
    pts_src = np.array([[464.48358, 727.9149], [2370.2458, 796.1351], [2310.8284, 2484.0332], [277.4284, 2525.8455]]).astype(np.float32)
    ## The above pts_src are those coordinates of the four windows that we want to map onto destination points
    pts_dst = np.array([[0, 0], [2000, 0], [2000, 2000], [0, 2000]]).astype(np.float32)
    ## The pts_dst are those points that have already been given to us and we want to use these two points to compute
    ## Homography Matrix
    
    # compute homography between src and dst
    # YOUR CODE HERE
    H,ones = cv2.findHomography(pts_src,pts_dst) # Gives us the homography matrix between the stitched image and
    # the final image to be computed!
    
    h1, w1, d1  = im_src.shape # Now because we have to do homography here, we have to make sure we define the limit 
    # points (corner points or bounds for the source image (stitched image in this case) and the destination image 
    # (image with perspective distortion removed))
    h2, w2, d2  = im_src.shape   # ONE ASSUMPTION MADE: The bounds of the stitched image and final image are assumed 
    # to be the same (not specified otherwise in the assignment!)
    
    src_corners = np.float32([[0,0],[0,h1],[w1,h1],[w1,0]]).reshape(-1,1,2) # computer corners based on the shape of 
    # images for source and destination!
    dst_corners = np.float32([[0,0],[0,h2],[w2,h2],[w2,0]]).reshape(-1,1,2)

    src_corners_warped = cv2.perspectiveTransform(src_corners,H) #Now we warp the stitched image coordinates using
    # the homography matrix computed at the start of THIS function! 
    
    # compute the bounds of the output rectified image (hint: cv2.perspectiveTransform)
    minimas = np.minimum(src_corners_warped,pts_dst) # to compute minimums from the destination 
    #image and the warped source image
    
    maximas = np.maximum(src_corners_warped,pts_dst) # to compute maximums from the destination 
    #image and the warped source image
        
    [xmin,xmax,ymin,ymax] = [np.min(minimas[:,:,0]), np.max(maximas[:,:,0]),np.min(minimas[:,:,1]),np.max(maximas[:,:,1])]
    
    # warp the image with bounds
    result = warp_image_with_bounds(im_src, H, [xmin, ymin, xmax, ymax]) # should output the final image with planar
    # homography done!
    

    # display
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))
    ax.imshow(result)
    ax.axis('off')
    
    return result  
```

```python
## int main()

im_src = plt.imread( 'a5-src.png' )
im_dst = plt.imread( 'a5-dst.png' )
stch   = stitch_images( im_src, im_dst )
print("stiched image computed!")
rect   = rectify(stch)
print("rectified image computed")
```

```python
# Displaying the two images side by side for comparison

plt.figure(figsize=(15,12))
plt.subplot(1,2,1)
plt.imshow(stch)
plt.title("Stitched Image with Perspective Distortion")

plt.subplot(1,2,2)
plt.imshow(rect)
plt.title("Final Image with Perspective Distortion Removed")
```
