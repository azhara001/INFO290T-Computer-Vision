- The convolution yields a blurry image (low pass filter)
- To sharpen the original image, you want to subtract the original image from the blur image (to get the high frequency component)
- based on an array (k), add the variable amounts of these blurs kl*(I-I_blur) back into the image
- Fix the contrast issue using skimage.exposure.match_histogram() to match the sharpened image's intensity distribution to the original image's intensity distribution

```python
path = "Einstein.png"
#I = cv2.imread(path)
#path = "Test.jpeg"
I = plt.imread(path)
rows,columns = I.shape[0],I.shape[1]
```

```python
# Trying with identity kernel 
identity_kernel = np.array([[0,0,0],[0,1,0],[0,0,0]])
identity_image = cv2.filter2D(I,ddepth=-1,kernel=identity_kernel)

#cv2.imshow('original',I)
#cv2.imshow('identity',identity_image)
#cv2.waitKey()
#cv2.destroyAllWindows()
```

```python
blur_kernel = [[1 / 256, 4  / 256,  6 / 256,  4 / 256, 1 / 256],
                   [4 / 256, 16 / 256, 24 / 256, 16 / 256, 4 / 256],
                   [6 / 256, 24 / 256, 36 / 256, 24 / 256, 6 / 256],
                   [4 / 256, 16 / 256, 24 / 256, 16 / 256, 4 / 256],
                   [1 / 256, 4  / 256,  6 / 256,  4 / 256, 1 / 256]]
```

```python
# Initializing kernels 
blur_kernel = np.array([[0.0625, 0.125, 0.0625],[0.125, 0.250, 0.125],[0.0625,0.125,0.0625]])
#blur_kernel = np.ones((7, 7), np.float32) / 49

I_blur = ndimage.convolve(I,blur_kernel,mode='reflect')

print(blur_kernel)
#I_blur = cv2.filter2D(I,ddepth=-1,kernel=blur_kernel)

plt.figure(figsize=[20,5])
plt.subplot(1,2,1)
plt.imshow(I_blur,cmap="gray",label="blurred image")
plt.title("blurred image")

plt.subplot(1,2,2)
plt.imshow(I,cmap='gray',label="original image")
plt.title("original image")
```

```python
print('convolution performed (I_blur)')
```

```python
sharp_kernel = np.array([[0.0091,-0.1134,0.0091],[-0.1134,1.4172,-0.1134],[0.0091,-0.1134,0.0091]])
I_sharp = ndimage.convolve(I,sharp_kernel,mode='reflect')
print('convolution performed (I_sharp)')
```

```python
k = np.arange(0,20,0.5)

I_sharps = []

for elem in k:
    sharpened_image = I+elem*(I-I_blur)
    sharpened_image = (match_histograms(image=sharpened_image,reference=I,channel_axis=None))
    I_sharps.append(sharpened_image)
    #imageio.imwrite(f"filename {elem}.jpeg", sharpened_image)
    plt.imsave(f"Sharpened Images/Image{elem}.jpeg",sharpened_image,cmap="gray")
    print(f"Image saved with k: {elem}")
    

print('images calculated!')
```

```python
# Checking for edge detection

horizontal_kernel = np.array([[-0.125, -0.250, -0.125],[0,0,0],[0.125,0.250,0.125]])
vertical_kernel = np.array([[-0.125,0,0.125],[-0.125,0,0.125],[-0.125,0,0.125]])

dx = ndimage.convolve(I,horizontal_kernel,mode='reflect')
dy = ndimage.convolve(I,vertical_kernel,mode='reflect')
dy_dx = np.sqrt(dx**2 + dy**2)
print('gradients computed')

plt.figure(figsize=[20,10])
plt.subplot(1,3,1)
plt.imshow(dx,cmap='gray')
plt.subplot(1,3,2)
plt.imshow(dy,cmap='gray')
plt.subplot(1,3,3)
plt.imshow(dy_dx,cmap='gray')
```
