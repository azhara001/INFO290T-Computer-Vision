### High Level Understanding of Assignment Part-1 (Sharpening)

- Import Image I in a numpy array 
- Figure out which blur kernel to use: (trying to understand the impact of kernel choice)<br>
    > blur_kernel = np.array([[0.0625, 0.125, 0.0625],[0.125, 0.250, 0.125],[0.0625,0.125,0.0625]])<br>
    > blur_kernel = np.ones((7, 7), np.float32) / 49 <br>
    > Gaussian blur
- The convolution yields a blurry image (low pass filter)
- To sharpen the original image, you want to subtract the original image from the blur image (to get the high frequency component)
- based on an array (k), add the variable amounts of these blurs kl*(I-I_blur) back into the image
- Fix the contrast issue using skimage.exposure.match_histogram() to match the sharpened image's intensity distribution to the original image's intensity distribution 
- Output the entire images as a GIF (GIF image output of these 40 images)

**Loading the image in a numpy array using plt.imread()**

```python
path = "Einstein.png"
#I = cv2.imread(path)
#path = "Test.jpeg"
I = plt.imread(path)
rows,columns = I.shape[0],I.shape[1]
```

**Initializing Blur Kernels and convolving with image**

Using two kernels for the sake of exploration on the impact of kernels on sharpening

```python
# Initializing kernels 

blur_kernels={} #blur kernels 
I_blur = {} #Dictionary of convolved images

blur_kernels['In class 3x3 kernel'] = np.array([[0.0625, 0.125, 0.0625],[0.125, 0.250, 0.125],[0.0625,0.125,0.0625]])
blur_kernels['7x7 constant values kernel'] = np.ones((7, 7), np.float32) / 49

I_blur['In class 3x3 kernel'] = ndimage.convolve(I,blur_kernels['In class 3x3 kernel'],mode='reflect')
I_blur['7x7 constant values kernel'] = ndimage.convolve(I,blur_kernels['7x7 constant values kernel'],mode='reflect')


# Using gaussian blur
sigma = 5.0
# apply Gaussian blur, creating a new image
I_blur['Gaussian Blur'] = skimage.filters.gaussian(I, sigma=(sigma, sigma), truncate=3.5, channel_axis=2)


plt.figure(figsize=[20,5])
plt.subplot(1,3,1)
plt.imshow(I_blur['In class 3x3 kernel'],cmap="gray")
plt.title("blurred image with 3x3 kernel as discussed in class")

plt.subplot(1,3,2)
plt.imshow(I_blur['7x7 constant values kernel'],cmap='gray',label="original image")
plt.title("7x7 constant values kernel")

plt.subplot(1,3,3)
plt.imshow(I_blur['Gaussian Blur'],cmap="gray")
plt.title("Gaussian Blur Kernel")
#I_blur = cv2.filter2D(I,ddepth=-1,kernel=blur_kernel)
```

**Adding High Frequency Component Back for sharpening of image**

Using the 3x3 kernel as discussed in class from this point on to sharpen the image

```python
k = np.arange(0,20,0.5)

I_sharps = []
I_blur = I_blur['In class 3x3 kernel']

for elem in k:
    sharpened_image = I+elem*(I-I_blur)
    sharpened_image = (match_histograms(image=sharpened_image,reference=I,channel_axis=None))
    I_sharps.append(sharpened_image)
    #imageio.imwrite(f"filename {elem}.jpeg", sharpened_image)
    plt.imsave(f"Sharpened Images/Image{elem}.jpeg",sharpened_image,cmap="gray")
    print(f"Image saved with k: {elem}")
    

print('images calculated!')
```

```python
io.mimsave('Einstein.gif', I_sharps)
```

### Part 2 of Assignment 

> Given a convolved signal, you want to deconvolve it linear algebraically using inverse matrix operation <br>

Ih : Gray Scale Convolved Image <br> 
h : Filter <br>
I : original image <br> 

> Equation: Ih = h*I 

**Loading the image in a numpy array using plt.imread()**

```python
path = "Einstein.png"
#I = cv2.imread(path)
#path = "Test.jpeg"
I = plt.imread(path)
rows,columns = I.shape[0],I.shape[1]
print(f"Rows: {rows}")
```

```python
filter_h = np.array([-1,0.001,1])
```

**Construction of filter matrix with shape 492x492**

```python
A = np.zeros([rows,columns])
h1 = A[0,:].copy()
h1[0:3] = filter_h

for row in range(rows):
    A[row,:] = np.roll(h1,shift=(row-1),axis=(0))
    
print(A)
```

**Deconvolve Function Declaration**

```python
def deconvolve(f_h,A):
    f_0 = np.dot(np.linalg.inv(A),f_h)
    f_0 = np.reshape(f_0,(1,rows))
    return f_0
```

```python
def convolution(im, h):
    
    # we can have the boundaries as zero in the output image
    im_h = np.zeros_like(im)
    
    # start and end positions
    if len(h.shape) < 2:
        sx = int(np.floor(h.shape[0]/2))
        sy = 0
    else:
        sy = int(np.floor(h.shape[0]/2))
        sx = int(np.floor(h.shape[1]/2))
    
    # use two for loops for convolution
    for i in range(im_h.shape[0]):
        for j in range(im_h.shape[1]):
            
            # wrap the pixel location around
            ys = np.arange(i-sy, i+sy+1)
            ys[ys < 0] = ys[ys < 0] + im_h.shape[0]
            ys = ys % im_h.shape[0]
            
            xs = np.arange(j-sx, j+sx+1)
            xs[xs < 0] = xs[xs < 0] + im_h.shape[1]
            xs = xs % im_h.shape[1]
                        
            # the convolved image
            im_h[i,j] = np.sum(h * im[ys, :][:, xs])
    
    return np.squeeze(im_h)
```

**Iterating over rows of image and deconvolving row-wise**

```python
print(I.shape)
print(A.shape)

Ih = convolution(I,filter_h)
plt.imshow(Ih,cmap="gray")
```

```python
deconvolved_image = np.zeros([rows,columns])

for row in range(rows):
    f_h =np.reshape(Ih[row,:],(rows,1))
    deconvolved_image[row,:] = deconvolve(f_h,A)
    
```

**Plotting the images outputs**

```python
plt.figure(figsize=[20,5])

plt.subplot(1,3,1)
plt.imshow(I,cmap="gray")
plt.title("Original Image I")

plt.subplot(1,3,2)
plt.imshow(Ih,cmap="gray")
plt.title("Convolved Image Ih")

plt.subplot(1,3,3)
plt.imshow(deconvolved_image,cmap="gray")
plt.title("De-Convolved Image Ih")
```
